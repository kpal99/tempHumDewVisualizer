```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
df = pd.read_csv('cpOutdoor_28-01-23.csv',
                 sep='\s+',
                 index_col='DateTime',
                 header=None,
                 parse_dates=False,
                 squeeze=True,
                 names=['Temperature', 'Humidity', 'DateTime'])
```

```python
df.index = pd.to_datetime(df.index, dayfirst=True)
```

```python
df['Temperature'] = np.where(df['Temperature'] > 100, df['Temperature'].mean(), df['Temperature'])
df['Humidity'] = np.where(df['Humidity'] > 100, df['Humidity'].mean(), df['Humidity'])
```

```python
df.plot(figsize=(20, 6))
plt.show()
```
